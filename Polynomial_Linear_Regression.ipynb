So, here we are going to look on the person's previous salary based on his position and his level of experience

```python
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns
```

# 1.Reading the Dataset

```python
dataset=pd.read_csv('Position_Salaries.csv')
X=dataset.iloc[:,1:-1].values
y=dataset.iloc[:,-1].values
```

# 2. Training the linear regression model on the dataset

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(X,y)
```

# 3. Lets work on creating Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2)

# creating matrix of features
X_poly=poly_reg.fit_transform(X)

#creating a new object , since we have already used the previous one to plot the linear regression model
lin_reg_two=LinearRegression()

lin_reg_two.fit(X_poly,y)
```

# 4. Visualizing the Linear Regression Model

```python
plt.scatter(X , y , color = 'red')

plt.plot(X , lin_reg.predict(X), color = 'blue')

plt.title('Truth or Bluff (Linear Regression)')

plt.xlabel('position level')

plt.ylabel('salary')

plt.show()
```

In the above graph, the red dots are the salaries

the line is for predicting the salaries.

As we can see its not a best fit line and the curve cant fit to the data. So lets check the same with polynomial_regression

```python
print(np.size(X_poly))

print(np.size(X))
```

# 5. Visualizing the Polynomial Regression results

```python
plt.scatter(X , y , color = 'red')

plt.plot(X, lin_reg_two.predict(X_poly), color = 'blue')

plt.title('Truth or Bluff (Polynomial Regression)')

plt.xlabel('Position Level')

plt.ylabel('Salary')

plt.show()
```

we have seen for both and in polynomial too, some points arent seem fitting the line, so we will try to plot for a smoother curve

--> how to do that? , by varying the degree during the data training.

first we have tried it with degree-2, now we will try it with degree - 4.

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)

# creating matrix of features
X_poly=poly_reg.fit_transform(X)

#creating a new object , since we have already used the previous one to plot the linear regression model
lin_reg_two=LinearRegression()

lin_reg_two.fit(X_poly,y)
```

```python
np.size(X_poly)
```

```python
plt.scatter ( X, y , color = 'red')
plt.plot(X,lin_reg_two.predict(X_poly),color='blue')

plt.title('Truth or bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

so to get more smoother curve, we are going to increase the distributions, instead of 1,2,3,4

we will make them as 1.1,1.2,1.3 ..... 9.1,9.2,9.3....


```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid= X_grid.reshape((len(X_grid),1))
plt.scatter(X,y,color='red')
plt.plot(X_grid,lin_reg_two.predict(poly_reg.fit_transform(X_grid)),color='blue')
plt.title('Truth or Bluff(SVR)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
